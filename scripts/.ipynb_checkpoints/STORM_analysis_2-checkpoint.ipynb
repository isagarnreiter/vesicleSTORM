```python
def normalize_array(arr):
    min_val = np.min(arr)
    max_val = np.max(arr)
    return (arr - min_val) / (max_val - min_val)

# apply the normalization function to the column of arrays
storm_data['nn_680_normalized'] = storm_data['nearest_neighbor_680'].apply(normalize_array)
```

```python
def ripley_k(array, volume):
    x = array[:,0]
    y = array[:,1]
    z = array[:,2]
    radii = list(np.linspace(20,1000,20))
    k = ripleyk.calculate_ripley(radii, volume*10**9, d1=x, d2=y, d3=z, CSR_Normalise=False)

    return k

storm_data['ripleyk'] = storm_data.apply(lambda row: ripley_k(row['points'], row['volume']), axis=1)
```

```python
storm_data['ripleyk'] = storm_data['ripleyk'].apply(np.array)
```

```python
x = '647nm'

dep_ripley = storm_data[storm_data['647nm'] == 'DEP647']['ripleyk'].to_numpy()
spon_ripley = storm_data[storm_data['647nm'] == 'SPON647']['ripleyk'].to_numpy()

csr = (np.linspace(20,1000,20)**3)*np.pi*4/3

plt.plot(np.array([np.mean(dep_ripley, axis=0), np.mean(spon_ripley, axis=0)]).T, alpha=0.5)


plt.xticks(list(np.linspace(0,20,6)), labels=list(np.linspace(20,1000,6)))
# Add titles and axis labels
plt.ylabel("ripley's K")
plt.xlabel('radius')
plt.legend(['dep', 'spon'])
plt.show()
# Show plot
```
